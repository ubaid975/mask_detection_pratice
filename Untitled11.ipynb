```python
zip=zipfile.ZipFile('covid-face-mask-detection-dataset.zip')
zip.extractall()
zip.close()
```

```python
train_path=r'/content/New Masks Dataset/Train'
```

```python
test_path=r'/content/New Masks Dataset/Test'
```

```python
validation_path='/content/New Masks Dataset/Validation'
```

```python
len(os.listdir(os.path.join(train_path,'Mask')))
```

```python
len(os.listdir(os.path.join(validation_path,'Mask')))
```

```python
len(os.listdir(os.path.join(test_path,'Mask')))
```

```python
gen=ImageDataGenerator(rescale=1/255,shear_range=0.2,zoom_range=0.2,
                             horizontal_flip=True,rotation_range=0.5)
```

```python
train_gen=gen.flow_from_directory(train_path,target_size=(224,224),batch_size=32)
```

```python
test_gen=gen.flow_from_directory(test_path,target_size=(224,224),batch_size=32,
                                  class_mode='binary')
```

```python
validation_gen=gen.flow_from_directory(validation_path,target_size=(224,224),batch_size=32,
                                  class_mode='binary')
```

```python
vgg19=VGG19(input_shape=(224,224,3),weights='imagenet')
```

```python
vgg19.summary()
```

```python
model=Sequential()
```

```python
for i in vgg19.layers[:-1]:
  model.add(i)
  model.trainable=False
```

```python
model.summary()
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.1),loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(train_gen,epochs=25,validation_data=validation_gen,
          )
```

```python
model.evaluate(test_gen)
```

```python
pred=np.where(model.predict(test_gen)>0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,precision_score,recall_score,log_loss
```

```python
pred.shape
```

```python
y_true=test_gen.classes
```

```python
confusion_matrix(y_true,pred)
```

```python
print(classification_report(y_true,pred))
```

```python
log_loss(y_true,pred)
```
